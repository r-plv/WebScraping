```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen
import numpy as np
import pandas as pd
```

```python
Name = []                            # List storing Product names
Rating = []                          # List storing Product Ratings
RAM_GB,ROM_GB = [],[]                # List storing RAM & ROM details of products
Display_in_cm = []                   # List storing Display information
RearCamera,FrontCamera = [],[]       # List storing Rear Camera & Front Camera specification
Battery = []                         # List for storing Battery specification
Processor = []                       # List of Processors of product
Price = []                           # List storing Price of products
```

```python
for i in range(1,7):
    url = 'https://www.flipkart.com/search?q=oneplus+mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page='+str(i)
    uClient = urlopen(url)
    html_page = uClient.read()
    uClient.close()
    soup_page = soup(html_page,'html.parser')
    
    for data in soup_page.findAll('div',{'class':'_3pLy-c row'}):
        names = data.find('div',attrs={'class':'_4rR01T'}).text
        ratings = data.find('div',attrs={'class':'_3LWZlK'})
        prices = int(data.find('div',attrs={'class':'_30jeq3 _1_WHN1'}).text.replace('â‚¹','').replace(',',''))
        details = data.find('div',attrs={'class':'fMghEO'})
        for each in details:
            detail = each.findAll('li',attrs={'class':'rgWa7D'})
            ram = int(detail[0].text.split()[0])
            rom = int(detail[0].text.split('|')[1].split()[0])
            display = float(detail[1].text.split()[0])
            rear_ = detail[2].text.split('|')[0].replace(' Rear Camera','')
            # Below code is used to handle 'list out of range' error
            # In some container 'Front Camera' info not exists.
            try:
                front_ = detail[2].text.split('|')[1].replace(' Front Camera','').rstrip('MP')
            except:
                front_ = np.nan
            battery = int(detail[3].text.split()[0])
            # In some container 'Processor' info not exists.
            try:
                if detail[4].text.split()[-1] == 'Processor':
                    processor_ = detail[4].text.rstrip(' Processor')
                else:
                    processor_ = np.nan
            except:
                processor_ = np.nan

        Name.append(names)
        RAM_GB.append(ram)
        ROM_GB.append(rom)
        Display_in_cm.append(display)
        RearCamera.append(rear_)
        FrontCamera.append(front_)
        Battery.append(battery)
        Price.append(prices)
        Processor.append(processor_)
        # There exists missing values in Rating section to handle those missing values we performed following code.
        if ratings==None:
            Rating.append(np.nan)
        else:
            Rating.append(float(ratings.text))
```

```python
len(Name),len(Rating),len(Price),len(RAM_GB),len(ROM_GB),len(Display_in_cm),len(RearCamera),len(FrontCamera),len(Battery),len(Processor)
```

```python
df = pd.DataFrame({'Name':Name,'RAM_GB':RAM_GB,'ROM_GB':ROM_GB,'Display_in_cm':Display_in_cm,'RearCamera':RearCamera,
                   'FrontCamera':FrontCamera,'Battery':Battery,'Processor':Processor,'Rating':Rating,'Price':Price})
```
