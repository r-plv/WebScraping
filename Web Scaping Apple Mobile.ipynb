```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen
import numpy as np
import pandas as pd
```

```python
Name = []                            # List storing Product names
Rating = []                          # List storing Product Ratings
ROM_GB = []                          # List storing ROM details of products
Display_in_cm = []                   # List storing Display information
RearCamera,FrontCamera = [],[]       # List storing Rear Camera & Front Camera specification
Processor = []                       # List of Processors of product
Price = []                           # List storing Price of products
```

```python
for i in range(1,19):
    url ='https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&p%5B%5D=facets.brand%255B%255D%3DAPPLE&page='+str(i)
    uClient = urlopen(url)
    html_page = uClient.read()
    uClient.close()
    soup_page = soup(html_page,'html.parser')
    
    for data in soup_page.findAll('div',{'class':'_3pLy-c row'}):
        names = data.find('div',attrs={'class':'_4rR01T'}).text
        ratings = data.find('div',attrs={'class':'_3LWZlK'})
        prices = int(data.find('div',attrs={'class':'_30jeq3 _1_WHN1'}).text.replace('₹','').replace(',',''))
        details = data.find('div',attrs={'class':'fMghEO'})
        for each in details:
            detail = each.findAll('li',attrs={'class':'rgWa7D'})
            rom = int(detail[0].text.split()[0])
            display = float(detail[1].text.split()[0])
            rear_ = detail[2].text.split('|')[0].replace(' Rear Camera','').replace('MP','')
            front_ = float(detail[2].text.split('| ')[1].replace(' Front Camera','').rstrip('MP'))
            processor_ = detail[3].text.rstrip(' Processor')

        Name.append(names)
        ROM_GB.append(rom)
        Display_in_cm.append(display)
        RearCamera.append(rear_)
        FrontCamera.append(front_)
        Price.append(prices)
        Processor.append(processor_)
        Rating.append(float(ratings.text))
```

```python
len(Name),len(Rating),len(Price),len(ROM_GB),len(Display_in_cm),len(RearCamera),len(FrontCamera),len(Processor)
```

```python
df = pd.DataFrame({'Name':Name,'ROM_GB':ROM_GB,'Display_in_cm':Display_in_cm,'RearCamera':RearCamera,
                   'FrontCamera':FrontCamera,'Processor':Processor,'Rating':Rating,'Price':Price})
```
